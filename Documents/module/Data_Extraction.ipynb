```python
import os
from zipfile import ZipFile

for file_name in range(len(id)):
    # Low resolution
    low_zip_file_path = os.path.join(low_file_path, id[file_name] + '.zip')
    low_zip_file_name = os.path.join(low_path, id[file_name])

    if not os.path.exists(low_zip_file_name) or not os.listdir(low_zip_file_name):
        os.makedirs(low_zip_file_name, exist_ok=True)
        with ZipFile(low_zip_file_path, 'r') as low_zip:
            low_zip.extractall(low_zip_file_name)
        print(f"Extracted: {low_zip_file_path}")
    else:
        print(f"Skipped extraction: {low_zip_file_name} already exists and is not empty.")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# 원본 데이터가 저장된 폴더 경로
data_dir = r"D:\project_Lockers\Data\datasets"

# 데이터가 분리되어 저장될 폴더 경로
output_dir = r"D:\project_Lockers\Data\split_data"

# 각 클래스 폴더를 순회하며 파일을 나눔
for class_dir in os.listdir(data_dir):
    class_path = os.path.join(data_dir, class_dir)
    
    if os.path.isdir(class_path):
        files = os.listdir(class_path)

        # 파일이 1700개인지 확인
        if len(files) == 1700:
            # 파일을 섞고 나누기
            train_files, temp_files = train_test_split(files, train_size=1000, random_state=1337)
            val_files, test_files = train_test_split(temp_files, test_size=350, random_state=1337)

            # 디렉토리 생성 및 파일 이동
            for dataset_type, dataset_files in zip(['train', 'val', 'test'], [train_files, val_files, test_files]):
                dataset_dir = os.path.join(output_dir, dataset_type, class_dir)
                os.makedirs(dataset_dir, exist_ok=True)
                for file_name in dataset_files:
                    shutil.move(os.path.join(class_path, file_name), os.path.join(dataset_dir, file_name))
        else:
            print(f"Skipping {class_dir} as it doesn't have exactly 1700 files.")
```
